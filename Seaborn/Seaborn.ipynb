# SEABORN
Seaborn is a library for making statistical graphics in Python. It is built on top of matplotlib and closely integrated with pandas data structures.

```python
#Importing dependencies

import seaborn as sns
%matplotlib inline
```

## DISTRIBUTION PLOTS
To visualize the distribution of the data set.

 - Distplot
 - Jointplot
 - Pairplot

```python
#Importing IRIS dataset for plotting

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### distplot
For the univariate analysis to show the distribution of obeservations.

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'])
```

### jointplot
For the bivariate analysis to compare the distplots of two variables.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

Kind:
 - Scatter
 - reg
 - resid
 - kde
 - hex

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

### pairplot
For the multivariate analysis of the entire dataframe to compare numerical columns and categorical columns.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

## MATRIX PLOTS
Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data

### heatmap

Analyzing the correlation with the heatmap.

```python
tips.corr()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

# CATEGORICAL PLOTS
To visualize and plot the categorical data.

 - countplot
 - barplot
 - boxplot

### countplot
Countplot is used to visualize the occurence of a given variable.

```python
sns.countplot(x='sex',data=tips)
```

### barplot
Barplot is used to visualize the categorical data using an aggregate funciton.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

### boxplot
Boxplots are used to visualize the distribution of categorical data.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution.
