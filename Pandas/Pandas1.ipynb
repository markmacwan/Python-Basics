# MISSING DATA

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)      #drops rows with nan >= 2
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())         #fills missing values with mean of column A
```

# GROUPBY

The groupby method allows you to group rows of data together and call aggregate functions

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

Groupby method to group rows based off of any column.

```python
df.groupby('Company')
```

```python
by_comp = df.groupby("Company")
```

```python
by_comp.mean()
```

```python
df.groupby('Company').mean()
```

Mathematical operations

```python
by_comp.std()   #Computes Standard Deviation of values
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()        #Computes descriptive statistics
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```

# Merging, Joining, Concatenating

Three main ways of combining a dataframe: Merging, Joining & Concatenating

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

## CONCATENATION

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

## MERGING

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

Merging uses same logic as merging SQL tables together.

```python
pd.merge(left,right,how='inner',on='key')
```

## JOINING

Joining combines two or more columns of different indexed dataframes.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 
```

```python
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
